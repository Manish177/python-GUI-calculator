```python
from tkinter import *
root= Tk()
root.title('calculator')
root.iconbitmap(r"C:\\Users\\MANISH\\Dropbox\\Untitled Folder 1\\calculator.ico")
def click(exp):
    global expression
    expression=expression+str(exp)
    myentry.delete(0,END)
    myentry.insert(0,expression)
    
def clear():
    global expression
    myentry.delete(0,END)
    expression=''
    
def equals():
    global expression
    result=str(eval(expression))
    myentry.delete(0,END)
    myentry.insert(0,result)
    expression=''     
    
myentry=Entry(root)   
myentry.grid(row=0, column=0, columnspan=3, ipadx=50, ipady=10)

mybutton1=Button(root, text='1', width=10, height=4, command=lambda:click('1'))
mybutton2=Button(root, text='2', width=10, height=4,command=lambda:click('2'))
mybutton3=Button(root, text='3', width=10, height=4,command=lambda:click('3'))
mybutton4=Button(root, text='4', width=10, height=4,command=lambda:click('4'))
mybutton5=Button(root, text='5', width=10, height=4,command=lambda:click('5'))
mybutton6=Button(root, text='6', width=10, height=4,command=lambda:click('6'))
mybutton7=Button(root, text='7', width=10, height=4,command=lambda:click('7'))
mybutton8=Button(root, text='8', width=10, height=4,command=lambda:click('8'))
mybutton9=Button(root, text='9', width=10, height=4,command=lambda:click('9'))
mybutton0=Button(root, text='0', width=10, height=4,command=lambda:click('0'))

mybutton_add=Button(root, text='+', width=10, height=4, command=lambda:click('+'))
mybutton_sub=Button(root, text='-', width=10, height=4,command=lambda:click('-'))
mybutton_dot=Button(root, text='.', width=10, height=4, command=lambda:click('.'))
mybutton_mult=Button(root, text='*', width=10, height=4,command=lambda:click('*'))
mybutton_div=Button(root, text='/', width=10, height=4,command=lambda:click('/'))
mybutton_clear=Button(root, text='Clear', width=22, height=4, command=lambda:clear())
mybutton_equals=Button(root, text='=', width=10, height=4, command=lambda: equals())


mybutton1.grid(row=3, column=0)
mybutton2.grid(row=3, column=1)
mybutton3.grid(row=3, column=2)
mybutton4.grid(row=2, column=0)
mybutton5.grid(row=2, column=1)
mybutton6.grid(row=2, column=2)
mybutton7.grid(row=1, column=0)
mybutton8.grid(row=1, column=1)
mybutton9.grid(row=1, column=2)
mybutton0.grid(row=4, column=0)

mybutton_add.grid(row=4, column=1)
mybutton_sub.grid(row=4, column=2)
mybutton_dot.grid(row=5, column=0)
mybutton_mult.grid(row=5, column=1)
mybutton_div.grid(row=5, column=2)
mybutton_clear.grid(row=6, column=0, columnspan=2)
mybutton_equals.grid(row=6, column=2)


root.mainloop()
```
